```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from warnings import filterwarnings
filterwarnings('ignore')
```

### Read the winequality.csv file and describe it.

```python
df = pd.read_csv("winequality.txt", sep = ",")
df.head(10)
```

```python
# Describe our data for each feature and use .info() for get information about our dataset
df.describe().T
```

```python
df.info()
```

```python
# Analyse missing values
df.isnull().sum()
```

```python
# Our label Distribution (countplot)
sns.countplot(x="quality", data = df);
```

```python
# Correlation and Histogram
sns.pairplot(df, hue = 'quality');
```

```python
df.corr()
```

```python
plt.figure(figsize= (10,10))
sns.heatmap(df.corr(), annot = True, cbar = True, vmin = -1, vmax = 1);
```

```python
# Check duplicated values and drop them
df.duplicated().sum()
```

```python
df[df.duplicated()]
```

```python
df.drop_duplicates(inplace = True)
df.reset_index(inplace = True)
df.drop('index', axis = 1, inplace = True)
df.head()
```

```python
# Created classes according to quality (if quality equals or higher than 6 this is highquality (class 1), if quality is lower than 6 that is lowquality (class 0) )
df["highquality"] = [1 if x >= 6 else 0 for x in df["quality"]]
df.head()
```

```python
# X and y splitting
X = df.drop(["quality", "highquality"], axis = 1)
y = df[["highquality"]]
```

```python
# Now we can see number of samples for each quality class
sns.countplot(x="highquality", data = y);
```

```python
# Outlier Detecting
# Isolation Forest
from sklearn.ensemble import IsolationForest
iso_model = IsolationForest(contamination=0.1)
predicted_iso = iso_model.fit_predict(X)
```

```python
# Delete outliers
not_outlier = predicted_iso != -1
not_outlier[0:10]
```

```python
new_X = X[not_outlier]
new_X.shape
```

```python
new_y =  y.loc[list(new_X.index)]
new_y.shape
```

```python
new_X.reset_index(inplace = True)
```

```python
new_X.drop('index', axis = 1, inplace = True)
```

```python
new_y.reset_index(inplace = True)
```

```python
new_y.drop('index', axis = 1, inplace = True)
```

```python
# Feature Scaling (We need scaled the features because some of them are different range and this makes problem for our models)

from sklearn.preprocessing import StandardScaler, MinMaxScaler
X_scaled = StandardScaler().fit_transform(new_X)
X_scaled[0:5]
```

```python
# Split Train and Test dataset. (0.7/0.3)
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_scaled, new_y, test_size = 0.30, random_state = 42)
print("X_train Shape: ", X_train.shape)
print("X_test Shape: ", X_test.shape)
print("y_train Shape: ", y_train.shape)
print("y_test Shape: ", y_test.shape)
```

```python
# Import ML models and metrics
from sklearn.model_selection import  GridSearchCV
from sklearn.metrics import confusion_matrix, accuracy_score,precision_score, recall_score, f1_score, classification_report
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
```

### DecisionTreeClassifier

```python
dtc_model = DecisionTreeClassifier()
dtc_params = {"max_depth": [2,3,4,6],
               "min_samples_split": [2,4,7,20,40,50],
               "criterion": ["gini", "entropy"],
               "max_features": [2,5,4,6]}

dtc_cv_model = GridSearchCV(dtc_model, dtc_params, cv = 5, n_jobs = -1, verbose = 2).fit(X_train, y_train)
```

```python
print("Optimum hyperparameter degerleri: ", "\n", dtc_cv_model.best_params_)
```

```python
tuned_dtc_model = DecisionTreeClassifier(criterion = 'gini',
                                         max_depth = 4,
                                         max_features = 6,
                                         min_samples_split = 4, random_state = 12).fit(X_train, y_train)
```

```python
# We can see that there is not an overfitting and underfitting problems
print("DecisionTreeClassifier Train:  ", tuned_dtc_model.score(X_train, y_train))
print("DecisionTreeClassifier Test:  ", tuned_dtc_model.score(X_test, y_test))
```

```python
# Feature Importance
Importance = pd.DataFrame({"Importance": tuned_dtc_model.feature_importances_},
                           index = pd.DataFrame(X_train, columns = X.columns).columns)

Importance.sort_values(by = 'Importance',
                       axis = 0,
                       ascending = True).plot(kind = 'barh',
                                             color = 'red')

plt.xlabel('Variable Importance')
plt.gca().legend_ = None
```

```python
# Generate Confusion Matrix 
y_pred = tuned_dtc_model.predict(X_test)
cm = confusion_matrix(y_test, y_pred)
ax = sns.heatmap(cm, annot = True, cbar = False, fmt = "g");
ax.set_xlabel('Predicted Labels',fontsize = 15)
ax.set_ylabel('True Labels',fontsize = 15)
plt.show()
```

```python
# accuracy, recall, precision ve f1-score

print("Accuracy = {}".format(accuracy_score(y_test, y_pred)))
print("Recall = {}".format(recall_score(y_test, y_pred, average='macro')))
print("Precision = {}".format(precision_score(y_test, y_pred, average='macro')))
print("F1 Score = {}".format(f1_score(y_test, y_pred,average='macro')))
```

### RandomForestClassifier

```python

rfc_model = RandomForestClassifier()

rfc_params = {"max_depth": [2, 5, 8, 10],
             "n_estimators": [100, 500, 1000],
             "max_features": [2,5,8],
             "min_samples_split": [2,5,10]}

rfc_cv_model = GridSearchCV(rfc_model, rfc_params, cv = 5, n_jobs = -1, verbose = 2).fit(X_train, y_train)
```

```python
print("Optimum hyperparameter degerleri: ", "\n", rfc_cv_model.best_params_)
```

```python
tuned_rfc_model = RandomForestClassifier(max_depth = 5,
                                         max_features = 5,
                                         min_samples_split = 5,
                                         n_estimators = 500, random_state = 99).fit(X_train, y_train)
```

```python
# We can see that there is not an overfitting and underfitting problems
print("RandomForestClassifier Train:  ", tuned_rfc_model.score(X_train, y_train))
print("RandomForestClassifier Test:  ", tuned_rfc_model.score(X_test, y_test))
```

```python
# Feature Importance
Importance = pd.DataFrame({"Importance": tuned_rfc_model.feature_importances_},
                           index = pd.DataFrame(X_train, columns = X.columns).columns)

Importance.sort_values(by = 'Importance',
                       axis = 0,
                       ascending = True).plot(kind = 'barh',
                                             color = 'red')

plt.xlabel('Variable Importance')
plt.gca().legend_ = None
```

```python
# Generate Confusion Matrix 
y_pred = tuned_rfc_model.predict(X_test)
cm = confusion_matrix(y_test, y_pred)
ax = sns.heatmap(cm, annot = True, cbar = False, fmt = "g");
ax.set_xlabel('Predicted Labels',fontsize = 15)
ax.set_ylabel('True Labels',fontsize = 15)
plt.show()
```

```python
# accuracy, recall, precision ve f1-score

print("Accuracy = {}".format(accuracy_score(y_test, y_pred)))
print("Recall = {}".format(recall_score(y_test, y_pred, average='macro')))
print("Precision = {}".format(precision_score(y_test, y_pred, average='macro')))
print("F1 Score = {}".format(f1_score(y_test, y_pred,average='macro')))
```

### GradientBoostingClassifier

```python
gbc_model = GradientBoostingClassifier()

gbc_params = {"learning_rate": [0.1, 0.01, 0.001, 0.05],
              "n_estimators": [10,190,500,1000,2000],
              "max_depth": range(3,10,2),
              "min_samples_split": [2, 5, 10]}
    
gbc_cv_model = GridSearchCV(gbc_model,
                            gbc_params,
                            cv = 5,
                            n_jobs = -1,
                            verbose = 2).fit(X_train, y_train)
```

```python
print("Optimum hyperparameter degerleri: ", "\n", gbc_cv_model.best_params_)
```

```python
tuned_gbc_model = GradientBoostingClassifier(learning_rate = 0.01,
                                             max_depth = 3,
                                             min_samples_split = 2,
                                             n_estimators = 190,
                                             random_state = 33).fit(X_train, y_train)
```

```python
# We can see that there is not an overfitting and underfitting problems
print("GradientBoostingClassifier Train:  ", tuned_gbc_model.score(X_train, y_train))
print("RandomForestClassifier Test:  ", tuned_gbc_model.score(X_test, y_test))
```

```python
# Feature Importance
Importance = pd.DataFrame({"Importance": tuned_gbc_model.feature_importances_},
                           index = pd.DataFrame(X_train, columns = X.columns).columns)

Importance.sort_values(by = 'Importance',
                       axis = 0,
                       ascending = True).plot(kind = 'barh',
                                             color = 'red')

plt.xlabel('Variable Importance')
plt.gca().legend_ = None
```

```python
# Generate Confusion Matrix 
y_pred = tuned_gbc_model.predict(X_test)
cm = confusion_matrix(y_test, y_pred)
ax = sns.heatmap(cm, annot = True, cbar = False, fmt = "g");
ax.set_xlabel('Predicted Labels',fontsize = 15)
ax.set_ylabel('True Labels',fontsize = 15)
plt.show()
```

```python
# accuracy, recall, precision ve f1-score

print("Accuracy = {}".format(accuracy_score(y_test, y_pred)))
print("Recall = {}".format(recall_score(y_test, y_pred, average='macro')))
print("Precision = {}".format(precision_score(y_test, y_pred, average='macro')))
print("F1 Score = {}".format(f1_score(y_test, y_pred,average='macro')))
```
